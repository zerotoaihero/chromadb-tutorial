**Import and get the chromadb client**

```python
import chromadb
chroma_client = chromadb.Client()
```

**Create a collection**

```python
collection = chroma_client.create_collection(name="emotion_analysis_collection_v1")
```

**Read and process emotion dataset**

```python
import pandas as pd
```

```python
df = pd.read_csv('emotion_sentimen_dataset.csv')
df
```

```python
import uuid
documents = []
metadatas = []
ids = []
documents = df['text'].head(500).tolist()
emotion_series = df['Emotion'].head(500)
emotion_list = [{"emotion": value} for value in emotion_series.tolist()]
emotion_metadatas = emotion_list
ids = df['index'].head(500).astype(str).tolist()
```

```python
emotion_metadatas
```

```python
collection.add(documents=documents,
    metadatas=emotion_metadatas,
    ids=ids)
```

```python
results = collection.query(
    query_texts=["I'm excited to reach home"],
    n_results=5
)
results
```

```python
chroma_client.delete_collection(name='emotion_analysis_collection_v1')
```
